```python
df1.pivot_table(values = "MaxTemp", index = ["Rainfall","Sunshine"], columns="Evaporation")
```

```python
plt.figure(figsize=(10,8))
plt.scatter(df1.MaxTemp,df1.Evaporation)
plt.xlabel("Evaporation")
plt.ylabel("Temp")
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(df1.corr())
```

```python
plt.figure(figsize = (20,10))
plt.boxplot(df1.Rainfall)
```

```python
plt.figure(figsize = (20,10))
sns.boxplot(x="Evaporation", y = "MaxTemp",data = df1)
```

```python
plt.figure(figsize = (20,10))
sns.stripplot(x= "WindGustDir", y ="Evaporation",data = df1)
```

```python
plt.figure(figsize = (20,10))
sns.swarmplot(x= "WindGustDir", y ="Evaporation",data = df1)
```

```python
sns.distplot(df1.Evaporation, kde = True, hist  = False)
```

```python
plt.figure(figsize = (20,10))
sns.factorplot(x= "Evaporation",y ="Rainfall",data = df1)
```

```python
plt.figure(figsize = (20,10))
sns.jointplot(df1.Sunshine,df1.Evaporation,kind = "kde")
```
