```python
plt.figure(figsize=(15,15))
sns.countplot(x = "askedFor", data = df.head(20),hue="deal")
```

```python
plt.figure(figsize=(15,15))
sns.distplot(df.askedFor,kde=True,hist=False)
```

```python
plt.figure(figsize=(15,15))
sns.jointplot(x= "askedFor",y = "exchangeForStake",data = df, kind= "scatter")
```

```python
plt.figure(figsize=(15,15))
sns.barplot(x= "askedFor",y = "valuation",data = df,hue= "deal")
```

```python
plt.figure(figsize=(15,15))
sns.boxplot(x = "exchangeForStake", y = "valuation",data = df,hue="deal")
```

```python
pd.pivot_table(df,values ="askedFor",index = ["valuation","deal"],columns= "exchangeForStake")
```

```python
plt.figure(figsize=(15,15))
sns.swarmplot(x= "exchangeForStake",y = "valuation",data = df, hue = "deal")
```
